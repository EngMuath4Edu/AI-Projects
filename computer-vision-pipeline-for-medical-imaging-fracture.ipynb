```python
# an X-ray image procressing tool to detect and highlight hairline fractures v 1
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image in grayscale
img = cv2.imread(r'YOUR_PATH\.jpg', 0) # change according to your file location on your computer 

# Check if the image is loaded successfully
if img is None:
    print("Error: Image not found or unable to load.")
    exit()

# Enhance contrast using histogram equalization
equalized_img = cv2.equalizeHist(img)

# Apply adaptive thresholding to highlight subtle fractures
adaptive_thresh = cv2.adaptiveThreshold(equalized_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Apply GaussianBlur to reduce noise and improve edge detection
blurred = cv2.GaussianBlur(equalized_img, (5, 5), 0)

# Apply Canny edge detection with adjusted thresholds for more sensitive detection
edges = cv2.Canny(blurred, 50, 150)

# Morphological operations to enhance edge features (dilate then erode to highlight fractures)
kernel = np.ones((3, 3), np.uint8)
dilated = cv2.dilate(edges, kernel, iterations=2)  # Increase dilation iterations for more prominent edges
eroded = cv2.erode(dilated, kernel, iterations=1)  # Erosion to refine the results

# Display the results: original, adaptive threshold, and processed edge images
plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(132)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Adaptive Thresholding')
plt.xticks([]), plt.yticks([])

plt.subplot(133)
plt.imshow(eroded, cmap='gray')
plt.title('Processed Edges')
plt.xticks([]), plt.yticks([])

plt.show()
```
